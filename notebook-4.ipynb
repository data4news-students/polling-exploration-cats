### Exploratory chart 1

```python
%%R

# 👉Chart 2
# Insert code for a ggplot chart below


ggplot(df) +
    aes(x=bias, y=partisan, color=bias>0, alpha=.05) +
    geom_point() +
    geom_vline(aes(xintercept=0), color='grey')
```

👉 Summary

_This chart investigates the extent to which partisan polls' bias lean towards their parties._

👉 Reflection

What I can conclude based on this chart:
* Democratic polls lean towards Democratic bias
* Republican polls lean towards Republican bias

What I can't conclude based on this chart:
